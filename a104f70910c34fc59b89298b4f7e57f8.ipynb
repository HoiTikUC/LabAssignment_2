```python
import pandas as pd
import numpy as np

# read the portland_park_trees.csv file with pandas
trees_pl = pd.read_csv('/arcgis/home/Assignment 2 Data/portland_park_trees.csv', sep=",", header=0 , encoding="ISO-8859-1")


## Q1


Quercus = trees_pl [ trees_pl ["Genus"] == "Quercus"]  ## filter out only Quercus from genus

Acer = trees_pl [ trees_pl ["Genus"] == "Acer"] ## filter out only Acer from genus

number_Quercus_Acer = (len(Quercus) + len(Acer)) ## count the number of Quercus and Acer

print (number_Quercus_Acer) ## print out the result



## Q2

Quercus_DBH50 = trees_pl [( trees_pl ["Genus"]=="Quercus") & ( trees_pl ["DBH"]>50)]  ## filter out trees that is Quercus and its DBH is higher then 50 at the same time
Acer_DBH50 = trees_pl [( trees_pl ["Genus"]=="Acer") & ( trees_pl ["DBH"]>50)]    ## filter out trees that is Acer and its DBH is higher then 50 at the same time

print (len(Acer_DBH50)+len(Quercus_DBH50)) ## print out the number of tree that is Quercus with DBH greater than 50 or Acer with DBH greater than 50




## Q3

Fraxinus = trees_pl [( trees_pl ["Genus"]=="Fraxinus")]  ## filter out only Fraxinus from genus

fraxinus_DBH_mean = Fraxinus [ "DBH" ].mean()  ## calculate the mean of Fraxius' DBH
Quercus_DBH_mean = Quercus [ "DBH" ].mean() ## calculate the mean of Quercus' DBH
Acer_DBH_mean = Acer [ "DBH" ].mean() ## calculate the mean of Acer's DBH

mean = [ fraxinus_DBH_mean, Quercus_DBH_mean, Acer_DBH_mean ] ## create a list of the three means

max_mean = max(mean) ## find out the max mean from the list

if max_mean == fraxinus_DBH_mean :  ## print out fraxinus if fraxinus is the max mean
    print ("fraxinus")

if max_mean == Quercus_DBH_mean :  ## print out Quercus if Quercus is the max mean
    print ("Quercus")

if max_mean == Acer_DBH_mean :  ## print out Acer if Acer is the max mean
    print ("Acer")   




## Q4

acer_species_count = trees_pl[trees_pl['Genus'] == 'Acer']['Genus_spec'].nunique() ## first fillter out tree with Acer Genus and count the number of distinct elements from Genus_spec

print (acer_species_count) ## print out the result





## Q5


cities_df = pd.read_csv('/arcgis/home/Assignment 2 Data/world_cities.csv', sep=",", header=0 , encoding="ISO-8859-1") ## create a dataframe with world_cities.csv

cities_df['pop_M'] = cities_df['pop'] / 1_000_000  ## create a new column of 'pop_M', which is pop divided by 1 million
cities_df.drop(columns=['pop'], inplace=True)   ## remove the pop cloumn from the dataframe


city_firstname = cities_df[cities_df['city'] == 'Craiova'] ## since my first name start with a c, the city of craiova is fillted out from the dataframe


country = city_firstname['country'].values[0] ## find out the country of craiova and named it 'country'


biggest_cities = cities_df[cities_df['country'] == country].nlargest(5, 'pop_M') ## by using pop_M, render the 5 largest city in the 'country' named earlier

print (biggest_cities) ## print out the result

```
